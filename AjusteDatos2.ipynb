```
import pandas as pd
doc="NivelAlta"
dataset=pd.read_excel('/content/'+doc+'.xlsx')
```

```
dataset['fecha_hora'] = pd.to_datetime(dataset['fecha_hora'])
dataset.head()
#df=dataset
```

```
#Agrupar los datos y promediarlos por hora
times = pd.DatetimeIndex(dataset.fecha_hora)
grouped = dataset.groupby([times.year,times.month,times.day,times.hour]).mean()
print (grouped)
```

```
grouped.to_excel(str(doc) + "_Hour.xlsx") 
```
